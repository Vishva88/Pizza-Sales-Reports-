```python
daily_trend = (
    df.groupby(['weekday_no', 'day_name'])['order_id']
      .nunique()
      .reset_index(name='Total_Orders')
      .sort_values('weekday_no')
)
daily_trend
plt.figure()
plt.plot(daily_trend['day_name'], daily_trend['Total_Orders'], marker='o')
plt.title('Daily Trend of Total Orders')
plt.xlabel('Day')
plt.ylabel('Total Orders')
plt.show()
```

```python
monthly_trend = (
    df.groupby(['month_no','month_name'])['order_id']
    .nunique()
    .reset_index(name='Total_Orders')
    .sort_values('month_no')
)

monthly_trend
plt.figure()
plt.bar(monthly_trend['month_name'], monthly_trend['Total_Orders'])
plt.title('Monthly Trend of Orders')
plt.xlabel('Month')
plt.ylabel('Total Orders')
plt.show()
```

```python
category_sales = (
    df.groupby('pizza_category')['total_price']
    .sum()
    .reset_index()
)

category_sales['PCT'] = round(
    category_sales['total_price'] * 100 / df['total_price'].sum(), 2
)

category_sales
plt.figure()
plt.pie(
    category_sales['total_price'],
    labels=category_sales['pizza_category'],
    autopct='%1.1f%%'
)
plt.title('Sales Distribution by Pizza Category')
plt.show()
```

```python
size_sales = (
    df.groupby('pizza_size')['total_price']
    .sum()
    .reset_index()
)

size_sales['PCT'] = round(
    size_sales['total_price'] * 100 / df['total_price'].sum(), 2
)

size_sales.sort_values('pizza_size')

plt.figure(figsize=(7,7))

plt.pie(
    size_sales['total_price'],
    labels=size_sales['pizza_size'],
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Sales Distribution by Pizza Size')
plt.tight_layout()
plt.show()
```

```python
feb_sales = df[df['order_date'].dt.month == 2]

category_quantity = (
    feb_sales.groupby('pizza_category')['quantity']
    .sum()
    .reset_index(name='Total_Quantity_Sold')
    .sort_values('Total_Quantity_Sold', ascending=False)
)

category_quantity
plt.figure(figsize=(10, 6))

plt.bar(
    category_quantity['pizza_category'],
    category_quantity['Total_Quantity_Sold']
)

plt.xlabel('Total Quantity Sold')
plt.ylabel('Pizza Category')
plt.title('Total Pizza Quantity Sold by Category (February)')

plt.tight_layout()
plt.show()
```

```python
top_revenue = (
    df.groupby('pizza_name')['total_price']
    .sum()
    .reset_index(name='Total_Revenue')
    .sort_values('Total_Revenue', ascending=False)
    .head(5)
)

top_revenue
plt.figure()

bars = plt.bar(
    top_revenue['pizza_name'],
    top_revenue['Total_Revenue']
)

plt.title('Top 5 Pizzas by Revenue')
plt.xlabel('Pizza Name',labelpad=15-20)
plt.xticks(rotation=30, ha='center')
plt.ylabel('Total Revenue')
plt.show()
```

```python
bottom_revenue = (
    df.groupby('pizza_name')['total_price']
    .sum()
    .reset_index(name='Total_Revenue')
    .sort_values('Total_Revenue')
    .head(5)
)

bottom_revenue
plt.figure(figsize=(12, 4))

plt.bar(
    bottom_revenue['pizza_name'],
    bottom_revenue['Total_Revenue']
)

plt.xlabel('Total Revenue')
plt.ylabel('Pizza Name')
plt.title('Bottom 5 Pizzas by Total Revenue')

plt.tight_layout()
plt.show()
```
